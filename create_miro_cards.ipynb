```python


def make_card(index, title, institute, themes, picture):
    width = 800
    height = 600
    d = draw.Drawing(width, height)

    #title = "my very long title"
    #themes = ['Thriving Planet', 'Living Environments']
    #d.append(draw.Rectangle(-80, -50, 40, 50, fill='#1248ff'))

    print(picture)
    
    if picture != None:
        d.append(draw.Image(0, 50, 800, 600, picture, embed=True))

    x = 0
    y = 0
    h = 30
    w = width/5

    colors = ['#ff595e', '#ffca3a', '#8ac926', '#1982c4', '#6a4c93']
    themes_names = ["Thriving Planet", "Living Environments", "Digital Future", "Health and Wellbeing", "Equal Society"]

    for i in range(5):
        if themes_names[i] in themes:
            d.append(draw.Rectangle(i*w, y, w, h, fill=colors[i]))
            tcolor = "black" if i==1 else "white"        
            d.append(draw.Text(themes_names[i], 16, i*w + w/2, y+h/2, fill=tcolor, font_family='Arial', text_anchor='middle', dominant_baseline='middle')) 
    
    d.append(draw.Rectangle(0, h, 100, h+40, fill='#00'))
    d.append(draw.Text(str(index).zfill(3), 40, 50, h+30, fill='white', font_family='Arial Black', text_anchor='middle', dominant_baseline='middle')) 
    d.append(draw.Rectangle(100, h, width, h+40, fill='#ffffff'))
    d.append(draw.Text(title, 40, 110, h+30, fill="black", font_family='Arial', dominant_baseline='middle'))     
    d.append(draw.Text(institute, 15, 110, h+60, fill="#aaaaaa", font_family='Arial', dominant_baseline='middle')) 

    return d


# x += width/5
# d.append(draw.Rectangle(x, 400, 800/5, 40, fill='#ffca3a'))

# x += width/5
# d.append(draw.Rectangle(x, 400, 800/5, 40, fill='#8ac926'))
# x += width/5
# d.append(draw.Rectangle(x, 400, 800/5, 40, fill='#1982c4'))
# x += width/5
# d.append(draw.Rectangle(x, 400, 800/5, 40, fill='#6a4c93'))


# d.append(draw.Text("thriving planet", 20, 20, 595, fill='white', font_family='Arial')) 
# d.append(draw.Text("living environment", 20, 20+160, 595, fill='white', font_family='Arial')) 
# d.append(draw.Text("digital future", 20, 20+2*160, 595, fill='white', font_family='Arial')) 
# d.append(draw.Text("health and wellbeing", 20, 20+3*160, 595, fill='white', font_family='Arial')) 
# d.append(draw.Text("equal society", 20, 20+4*160, 595, fill='white', font_family='Arial')) 

d = make_card(102, "my long title", "tudelft", ['Thriving Planet', 'Equal Society'], "000002-R-img-5164.jpg")

d

```
